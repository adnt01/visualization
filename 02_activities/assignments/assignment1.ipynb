```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

colors = ["#9b59b6", "#3498db"]

# Create data frame with randomly selected x and y positions
df = pd.DataFrame(np.random.random((100, 2)), columns=["x", "y"])

# Add a column: the color depends on x and y values, but you can use any function you want
is_selected_dot = (df["x"] > 0.6) & (df["y"] > 0.6)
df["color"] = np.where(is_selected_dot, colors[0], colors[1])

# plot
sns.set_style("darkgrid")
sns.scatterplot(
    data=df,
    x="x",
    y="y",
    color=df["color"],
    s=400,
    alpha=0.8,
)

plt.show()
```

### Data Visualization: Subplots and Combining Visualizations

```python
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(9, 3))

font1 = {'family':'serif','color':'indigo'}
font2 = {'family':'serif','color':'red'}
ax1.scatter(x1,y1, color = 'blue')
ax2.bar(x2,y2, color ='indigo')

ax1.set_title('Scatter Plot', fontdict = font1)
ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
ax2.set_title('Bar Plot', fontdict = font2)
fig.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4)) 
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
```

```python
tips = sns.load_dataset("tips")
print(tips)
tips.head()
```

```python
sns.set_style('whitegrid')

sns.lineplot(data=tips,       # choose our dataset
           x='total_bill',  # define our x variable
           y='tip')         # define our y variable
```

```python
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')
tipgraph.set(title='Tips vs. Total Bill', 
            xlabel='Total Bill ($)', 
            ylabel='Tip Amount ($)')
```

```python

fig = plt.subplots(figsize=(10, 3))
tipgraph = sns.lineplot(data=tips,
                      x='total_bill',
                      y='tip',
                      color = 'hotpink',
                      linestyle = '--',
                      linewidth = 3,
                      marker = 'o',
                      markerfacecolor = 'indigo')
```

```python
tipgraph = sns.scatterplot(data=tips, x='total_bill',
                        y='tip', style = 'time', hue = 
                        'day', palette = ['purple',
                        'hotpink', 'deepskyblue',
                        'yellowgreen'])
tipgraph.set(title='Tips vs. Total Bill',
            xlabel='Total Bill ($)',
            ylabel='Tip Amount ($)')
```

```python
sns.pairplot(
      data = tips,
      hue = 'day')
```

```python
daysplot = sns.relplot(
            data=tips,
            x="total_bill",
            y="tip",
            hue="sex",
            col="day",
            kind="scatter",
            col_wrap=2)
```

```python
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines='skip')
df
```

```python
text = " ".join(each for each in df.quote)
print(text)
wordcloud = WordCloud(background_color="white",
                      colormap = 'inferno',
                      random_state = 613).generate(text)

fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud)
ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2([A, B],
                            set_labels = ('Fruits', 'Pies'),
                            set_colors=("blue", "red"),
                            alpha=0.5)
plt.show()
```

```python
diagram = venn2([A, B],
                            set_labels = ('Fruits', 'Pies'),
                            set_colors=("blue", "red"),
                            alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B)) #left
diagram.get_label_by_id("11").set_text("\n".join(A & B)) #middle
diagram.get_label_by_id("01").set_text("\n".join(B - A))

plt.show()
```
